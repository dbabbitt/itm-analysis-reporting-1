```python

# Get all CSVs into one data frame
if nu.pickle_exists('frvrs_logs_df'):
    frvrs_logs_df = nu.load_object('frvrs_logs_df')
    print(frvrs_logs_df.shape)
    # df = frvrs_logs_df.sample(4).dropna(axis='columns', how='all')
    # display(df.T)
```

```python

mask_series = frvrs_logs_df.columns.map(lambda x: 'require' in x)
columns_list = frvrs_logs_df.columns[mask_series].tolist()
frvrs_logs_df[columns_list].drop_duplicates().sort_values(columns_list)
```

```python

# Filter the logs data frame to only include rows where the required procedure column is not null and not equal to "none"
mask_series = ~frvrs_logs_df.injury_record_required_procedure.isnull() & (frvrs_logs_df.injury_record_required_procedure != 'none')
df = frvrs_logs_df[mask_series]

# Display a sample of the filtered data frame, dropping any columns that are all null

# Display a sample of the data frame, dropping columns with all NaN values and transposing it
display(df.sample(min(4, df.shape[0])).dropna(axis='columns', how='all').T)

# Create a list of the columns to group by



# Iterate over the grouped data frame
for (session_uuid, scene_id, patient_id), patient_df in df.groupby(fu.patient_groupby_columns):

    # Update the base mask series with the groupings
    base_mask_series = True
    for cn in fu.patient_groupby_columns: base_mask_series &= (frvrs_logs_df[cn] == eval(cn))

    # Only rows with an elapsed time greater than or equal to the maximum elapsed time of the group are included
    base_mask_series &= (frvrs_logs_df.action_tick >= patient_df.action_tick.max())

    # Filter the frvrs_logs_df data frame using the base mask series
    df2 = frvrs_logs_df[base_mask_series]

    # Display a sample of the filtered data frame, dropping any columns that are all null and sorting the data frame by elapsed time
    display(df2.sample(min(4, df2.shape[0])).dropna(axis='columns', how='all').sort_values('action_tick').T)

    # Break the loop after the first iteration
    break
```
